```python
import tensorflow as tf
import pandas as pd
pd.set_option('display.max_rows', 15)
```

```python
class dataset:
    risk_dataset = pd.read_csv('dataset.csv')
```

```python
dataset.risk_dataset
```

```python
class one_hot:
    one_hot_encoding = pd.get_dummies(dataset.risk_dataset) #get_dummies is used to convert categorical values to numerical values
```

```python
one_hot.one_hot_encoding
```

```python
train = one_hot.one_hot_encoding.sample(frac=0.8)#Returns a random sample of items from an axis of object
test = one_hot.one_hot_encoding.drop(train.index)
```

```python
train_input = train.filter(['Probability', 'Impact'])
train_label = train.filter(['Label_High', 'Label_Low', 'Label_Moderate'])
test_input = test.filter(['Probability', 'Impact'])
test_label = test.filter(['Label_High', 'Label_Low', 'Label_Moderate'])
```

```python
X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 3])
W = tf.Variable(tf.random_normal([2, 3]), name='weight')
b = tf.Variable(tf.random_normal([3]), name='bias')
```

```python
logits = tf.nn.softmax(tf.matmul(X,W)+b)
hypothesis = tf.nn.softmax(logits)
```

```python
cost = tf.reduce_mean(-tf.reduce_sum(Y * tf.log(logits), axis=1))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-1).minimize(cost)

prediction = tf.argmax(hypothesis, 1)
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(Y,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range (20001):
        sess.run(optimizer, feed_dict={X: train_input, Y: train_label})
        if step % 200 == 0:
            loss, acc = sess.run([cost, accuracy], feed_dict={X: train_input, Y: train_label})
            print("Step: {:5}\tLoss: {:.3f}\tACC: {:.2%}".format(step, loss, acc))
```

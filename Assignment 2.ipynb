```python
import tensorflow as tf
import numpy as np #to use np.Loadtxt

#1st step: Bring / Define the dataset
xy = np.loadtxt('diabetes-data.csv', delimiter=',', dtype=np.float32)
x_data = xy[:,0:-1]
y_data = xy[:, [-1]]

#2nd step: placeholders
X = tf.placeholder(tf.float32, shape=[None , 8])
Y = tf.placeholder(tf.float32, shape=[None , 1])

W = tf.Variable(tf.random_normal([8, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

#3rd step: Define a hypothesis
hypothesis = tf.sigmoid(tf.matmul(X, W) + b)

#4th step: Cost funciton
cost = -tf.reduce_mean(Y*tf.log(hypothesis)+(1-Y)*tf.log(1-hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate=1e-1).minimize(cost)

#5th step: Prediction
prediction = tf.cast(hypothesis>0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(prediction, Y), dtype=tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(1001):
        cost_val, _ = sess.run([cost, train], feed_dict={X: x_data, Y: y_data})
        h, p, a = sess.run([hypothesis, prediction, accuracy], feed_dict={X: x_data, Y: y_data})
        
        if step%20==0:
            print('\nHypothesis: ', h, '\nPrediction', p, '\nAccuracy: ', a)
```
